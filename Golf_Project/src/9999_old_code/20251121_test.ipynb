```python
from pathlib import Path
import re

IMG_DIR = Path("/home/dw/ws_job_msislab/Golf_Project/data/for_study/20251024_merge_data/20250930")
IMG_EXT = {".jpg", ".jpeg", ".png" , ".txt"}

count = 0

pattern = re.compile("bg")

for img_path in IMG_DIR.iterdir():
    if img_path.suffix.lower() in IMG_EXT:
        name = img_path.stem
        
        # 'bg' 등장 횟수 카운트 (정규식)
        if len(pattern.findall(name)) >= 2:
            print(f"삭제: {img_path.name}")
            img_path.unlink()
            count += 1

print(f"\n총 삭제된 파일: {count}개")

```

```python
from pathlib import Path

# 검사할 폴더
BASE = Path("/home/dw/ws_job_msislab/Golf_Project/data/for_study/20251024_merge_data/20250721")

IMG_EXT = {".jpg", ".jpeg", ".png"}
LABEL_EXT = ".txt"

images = []
labels = []

# 파일 수집
for f in BASE.iterdir():
    if f.suffix.lower() in IMG_EXT:
        images.append(f.stem)
    elif f.suffix.lower() == LABEL_EXT:
        labels.append(f.stem)

images = set(images)
labels = set(labels)

# 매칭 실패 항목 검사
no_label = images - labels      # 이미지 있는데 라벨 없음
no_image = labels - images      # 라벨 있는데 이미지 없음

print("======== 결과 ========")
print(f"총 이미지 개수: {len(images)}")
print(f"총 라벨 개수  : {len(labels)}")
print()

print(f"라벨 없는 이미지: {len(no_label)}개")
for x in sorted(no_label):
    print("  -", x)

print()
print(f"이미지 없는 라벨: {len(no_image)}개")
for x in sorted(no_image):
    print("  -", x)

print()
matched = len(images.intersection(labels))
print(f"정상 매칭된 개수: {matched}")
print(f"매칭률: {matched / max(len(images), 1) * 100:.2f}%")
```
