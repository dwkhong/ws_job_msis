```python
from pathlib import Path
import cv2

bad_dir = Path("/home/dw/ws_job_msislab/Golf_Project/data/20250721/20250721_bad_data")

IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".webp",
            ".JPG", ".JPEG", ".PNG", ".BMP", ".WEBP"}

deleted = 0
for img in bad_dir.rglob("*"):
    if img.suffix not in IMG_EXTS:
        continue
    # cv2가 정상적으로 읽을 수 있는지 체크
    try:
        im = cv2.imread(str(img))
        if im is None:
            raise ValueError("Cannot open")
    except Exception:
        img.unlink()
        deleted += 1
        print(f"삭제: {img}")

print(f"\n총 {deleted} 개의 손상된 이미지 삭제 완료.")
```

```python
from pathlib import Path
import os

# ===== 대상 폴더 =====
folder = Path("/home/dw/ws_job_msislab/Golf_Project/data/20250929/20250929_bad_data")

# ===== 파일들 정렬 후 순차 이름 변경 =====
files = sorted(folder.iterdir())  # 이름순 정렬
for i, f in enumerate(files, start=1):
    if f.is_file():
        ext = f.suffix  # 확장자 (.jpg, .png 등)
        new_name = f"20250929_{i:05d}{ext}"  # 예: 20250929_00001.jpg
        new_path = folder / new_name
        f.rename(new_path)
        print(f"{f.name} -> {new_name}")
```

```python
import os
from pathlib import Path

# 기준 경로
BASE = Path("/home/dw/ws_job_msislab/Golf_Project/data/for_study/20250929_merge_data_copy_only_0904_data")

# 대상 서브폴더 (images와 labels 각각의 test/train/val)
TARGETS = [
    BASE / "images/test", BASE / "images/train", BASE / "images/val",
    BASE / "labels/test", BASE / "labels/train", BASE / "labels/val",
]

# 삭제 조건
KEYWORDS = ["20250721", "20250725"]
EXTS = [".jpg", ".txt"]

# 실행
for folder in TARGETS:
    if not folder.exists():
        continue
    for file in folder.glob("*"):
        if file.suffix.lower() in EXTS and any(key in file.name for key in KEYWORDS):
            print(f"삭제됨: {file}")
            file.unlink()
```

```python
import os

folder = "/home/dw/ws_job_msislab/Golf_Project/data/20250930/20250930_bad_data"
files = sorted(os.listdir(folder))  # 정렬된 순서대로
pad = 5  # 자릿수 (00001 ~)

for i, fname in enumerate(files, start=1):
    old_path = os.path.join(folder, fname)
    # 확장자 보존
    ext = os.path.splitext(fname)[1]
    new_name = f"20250930_{i:0{pad}d}{ext}"
    new_path = os.path.join(folder, new_name)
    os.rename(old_path, new_path)
    print(f"{fname} -> {new_name}")
```
