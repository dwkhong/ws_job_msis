```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
YOLO í´ë” ë§¤ì¹­ ì ê²€ê¸° (ì½˜ì†” ì¶œë ¥ ì „ìš©)
- images/{train,val,test}, labels/{train,val,test} êµ¬ì¡° ê²€ì‚¬
- JPG ì—†ëŠ” TXT, TXT ì—†ëŠ” JPG, ì¤‘ë³µ í™•ì¥ì ëª¨ë‘ í™•ì¸
"""

from pathlib import Path

# ====== ì„¤ì • ======
BASE_DIR = Path("/home/dw/ws_job_msislab/Golf_Project/data/for_test/test_20251113")
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".webp", ".JPG", ".JPEG", ".PNG", ".BMP", ".WEBP"}
LABEL_EXT = ".txt"
SPLITS = ["train", "val", "test"]

def load_files(img_dir, lbl_dir):
    images_by_stem = {}
    labels_by_stem = {}
    stems_to_exts = {}

    for p in img_dir.rglob("*"):
        if p.is_file() and p.suffix in IMG_EXTS:
            stem = p.stem
            images_by_stem.setdefault(stem, p)
            stems_to_exts.setdefault(stem, set()).add(p.suffix.lower())

    for p in lbl_dir.rglob("*"):
        if p.is_file() and p.suffix == LABEL_EXT and p.name != "classes.txt":
            labels_by_stem[p.stem] = p

    return images_by_stem, labels_by_stem, stems_to_exts

def main():
    print(f"\nğŸ“‚ ê²€ì‚¬ ëŒ€ìƒ: {BASE_DIR}\n")
    for split in SPLITS:
        img_dir = BASE_DIR / "images" / split
        lbl_dir = BASE_DIR / "labels" / split

        if not img_dir.exists() or not lbl_dir.exists():
            print(f"âš ï¸ {split}: í´ë”ê°€ ì¡´ì¬í•˜ì§€ ì•ŠìŒ (images={img_dir.exists()}, labels={lbl_dir.exists()})")
            continue

        images_by_stem, labels_by_stem, stems_to_exts = load_files(img_dir, lbl_dir)

        all_stems = set(images_by_stem) | set(labels_by_stem)
        unmatched_imgs, unmatched_labels, matched = [], [], []
        duplicate_exts = []

        for stem in sorted(all_stems):
            img = images_by_stem.get(stem)
            lbl = labels_by_stem.get(stem)
            if img and lbl:
                matched.append(stem)
            elif img and not lbl:
                unmatched_imgs.append(stem)
            elif lbl and not img:
                unmatched_labels.append(stem)

        for stem, exts in stems_to_exts.items():
            if len(exts) > 1:
                duplicate_exts.append((stem, exts))

        print(f"====================== {split.upper()} ======================")
        print(f"âœ… ì •ìƒ ë§¤ì¹­: {len(matched)}ê°œ")
        print(f"âš ï¸ ì´ë¯¸ì§€ë§Œ ìˆê³  ë¼ë²¨ ì—†ìŒ: {len(unmatched_imgs)}ê°œ")
        print(f"âš ï¸ ë¼ë²¨ë§Œ ìˆê³  ì´ë¯¸ì§€ ì—†ìŒ: {len(unmatched_labels)}ê°œ")
        print(f"âš ï¸ ì¤‘ë³µ í™•ì¥ì: {len(duplicate_exts)}ê°œ")

        if unmatched_imgs:
            print("\nğŸ“¸ ì´ë¯¸ì§€ë§Œ ì¡´ì¬í•˜ëŠ” íŒŒì¼ë“¤:")
            for s in unmatched_imgs[:10]:
                print("  -", s)
            if len(unmatched_imgs) > 10:
                print(f"  ... (ì´ {len(unmatched_imgs)}ê°œ)\n")

        if unmatched_labels:
            print("\nğŸ“ ë¼ë²¨ë§Œ ì¡´ì¬í•˜ëŠ” íŒŒì¼ë“¤:")
            for s in unmatched_labels[:10]:
                print("  -", s)
            if len(unmatched_labels) > 10:
                print(f"  ... (ì´ {len(unmatched_labels)}ê°œ)\n")

        if duplicate_exts:
            print("\nâš ï¸ ì¤‘ë³µ ì´ë¯¸ì§€ í™•ì¥ì ì¡´ì¬:")
            for stem, exts in duplicate_exts[:10]:
                print(f"  - {stem} ({', '.join(exts)})")
            if len(duplicate_exts) > 10:
                print(f"  ... (ì´ {len(duplicate_exts)}ê°œ)\n")

        print("-------------------------------------------------------------")

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
============================================================
ğŸ¨ YOLO ë°ì´í„° ìƒ‰ìƒ ì¦ê°•(+í™•ë¥  ì¢Œìš°ë°˜ì „) â€” testë§Œ ì¦ê°• + ì „ì²´ í´ë” & data.yaml ìƒì„±
 - ì›ë³¸:  test_YYYYMMDD/{images,labels}/{train,val,test}
 - ê²°ê³¼:  test_YYYYMMDD_color/{images,labels}/{train,val,test}
   Â· train/val : ì›ë³¸ ê·¸ëŒ€ë¡œ í¬í•¨(ê¸°ë³¸: symlink, ì‹¤íŒ¨ ì‹œ copy)
   Â· test      : 
        - ì›ë³¸ ì´ë¯¸ì§€/ë¼ë²¨ ê·¸ëŒ€ë¡œ ë³µì‚¬
        - + ìƒ‰ìƒ ë³€í™˜ + (í™•ë¥ )ì¢Œìš°ë°˜ì „(ë¼ë²¨ ë™ê¸°í™”)ëœ ì¦ê°•ë³¸ ì¶”ê°€
        - íŒŒì¼ëª…: "xxx.jpg" (ì›ë³¸), "xxx_color(.jpg|.png...)", "xxx_color_flip..."
        - ì •ë ¬ ì‹œ ì›ë³¸ â†’ ì¦ê°• ìˆœìœ¼ë¡œ ë³´ì´ê²Œ êµ¬ì„±
 - ë¼ë²¨ ë³€í™˜: ì¢Œìš°ë°˜ì „ ì‹œ x_center -> (1 - x_center) (Albumentationsë¡œ ìë™ ì²˜ë¦¬)
 - data.yaml ìƒì„±: train/val/test ëª¨ë‘ ë§ì¶°ì„œ ê¸°ë¡
 - âœ… ë¼ë²¨ ì—†ëŠ” BG ì´ë¯¸ì§€ë„ ì¦ê°•/ì €ì¥ (ë¹ˆ ë¼ë²¨ ìƒì„±)
 - âœ… test split: ì›ë³¸ + ì¦ê°• í•©ì´ MAX_TEST_IMAGES(ê¸°ë³¸ 2000)ê¹Œì§€, ë””ë´‡(í´ë˜ìŠ¤ 0) ìœ„ì£¼ë¡œ ì¦ê°•
============================================================
"""

from pathlib import Path
import shutil
import cv2
import os
import random
import numpy as np
import albumentations as A

# ===== ì‚¬ìš©ì ì„¤ì • =====
BASE_DIR = Path("/home/dw/ws_job_msislab/Golf_Project/data/for_test/test_20251113")
OUT_DIR  = BASE_DIR.parent / (BASE_DIR.name + "_c")   # /.../test_20251112_color
IMG_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".webp", ".JPG", ".JPEG", ".PNG", ".BMP", ".WEBP"}

# YOLO í´ë˜ìŠ¤ (í•„ìš” ì‹œ ìˆ˜ì •)
YOLO_NAMES = [
    "Divot", "Fixed_Divot", "Diseased_Grass", "Confused_Object",
    "Pole", "Sprinkler", "Drain", "Golf_ball"
]

USE_SYMLINKS     = True        # True: train/valì„ ì‹¬ë³¼ë¦­ ë§í¬ë¡œ í¬í•¨ (ë¹ ë¦„). ë§í¬ ë¶ˆê°€ ì‹œ ìë™ ë³µì‚¬.
RANDOM_SEED      = 33
HFLIP_PROB       = 0.0         # ì¢Œìš°ë°˜ì „ í™•ë¥  (0.0 ~ 1.0)
MAX_TEST_IMAGES  = 2000        # âœ… test split: ì›ë³¸ + ì¦ê°• í•© ëª©í‘œ ê°œìˆ˜

random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

# ===== ê²½ë¡œ ì •ì˜ =====
SRC_IMG = {s: BASE_DIR / f"images/{s}" for s in ("train", "val", "test")}
SRC_LBL = {s: BASE_DIR / f"labels/{s}" for s in ("train", "val", "test")}
DST_IMG = {s: OUT_DIR  / f"images/{s}" for s in ("train", "val", "test")}
DST_LBL = {s: OUT_DIR  / f"labels/{s}" for s in ("train", "val", "test")}

# ===== ë””ë ‰í† ë¦¬ ìœ í‹¸ =====
def ensure_dirs():
    for split in ("train", "val", "test"):
        DST_IMG[split].mkdir(parents=True, exist_ok=True)
        DST_LBL[split].mkdir(parents=True, exist_ok=True)

def safe_symlink(src: Path, dst: Path):
    try:
        if dst.exists() or dst.is_symlink():
            if dst.is_dir():
                shutil.rmtree(dst)
                dst.mkdir(parents=True, exist_ok=True)
            else:
                dst.unlink()
        os.symlink(src, dst)
        return True
    except Exception:
        return False

def copy_dir_tree_images_labels(src_img: Path, src_lbl: Path, dst_img: Path, dst_lbl: Path):
    n_img = n_lbl = 0
    if src_img.exists():
        for p in src_img.iterdir():
            if p.suffix.lower() in IMG_EXTS:
                shutil.copy2(p, dst_img / p.name)
                n_img += 1
    if src_lbl.exists():
        for p in src_lbl.iterdir():
            if p.suffix.lower() == ".txt":
                shutil.copy2(p, dst_lbl / p.name)
                n_lbl += 1
    return n_img, n_lbl

# ===== ìƒ‰ìƒ ì¦ê°• ì •ì˜ (ê¸°í•˜ ë³€í˜• ì—†ìŒ) =====
def _build_color_ops():
    return {
        "gamma_bright": A.RandomGamma(gamma_limit=(55, 85), p=1.0),    # ë” ë°ê²Œ
        "gamma_dark":   A.RandomGamma(gamma_limit=(120, 180), p=1.0),  # ë” ì–´ë‘¡ê²Œ
        "warm": A.Compose([
            A.ColorJitter(brightness=0.10, contrast=0.15, saturation=0.08, hue=0.015, p=1.0),
            A.RGBShift(r_shift_limit=(8, 20), g_shift_limit=(-4, 6), b_shift_limit=(-20, -8), p=0.9),
            A.RandomGamma(gamma_limit=(95, 120), p=1.0),
        ], p=1.0),
        "cool": A.Compose([
            A.ColorJitter(brightness=0.08, contrast=0.12, saturation=0.06, hue=0.015, p=1.0),
            A.RGBShift(r_shift_limit=(-20, -8), g_shift_limit=(-6, 6), b_shift_limit=(8, 20), p=0.9),
            A.RandomGamma(gamma_limit=(95, 120), p=1.0),
        ], p=1.0),
        "desat": A.Compose([
            A.HueSaturationValue(hue_shift_limit=0, sat_shift_limit=(-45, -25), val_shift_limit=(-5, 10), p=1.0),
            A.RandomGamma(gamma_limit=(90, 115), p=1.0),
        ], p=1.0),
        "satboost": A.Compose([
            A.HueSaturationValue(hue_shift_limit=0, sat_shift_limit=(20, 45), val_shift_limit=(-5, 10), p=1.0),
            A.RandomGamma(gamma_limit=(90, 115), p=1.0),
        ], p=1.0),
        "identity": A.Compose([], p=1.0),
    }

COLOR_OPS = _build_color_ops()
WEIGHTS = {
    "gamma_bright": 1.0,
    "gamma_dark":   1.0,
    "warm":         1.0,
    "cool":         1.0,
    "desat":        0.8,
    "satboost":     0.8,
    "identity":     0.3,
}

def build_color_aug(weights: dict) -> A.OneOf:
    choices = []
    for name, t in COLOR_OPS.items():
        w = float(weights.get(name, 0.0))
        if w > 0:
            t.p = w
            choices.append(t)
    if not choices:
        choices = [A.NoOp(p=1.0)]
    return A.OneOf(choices, p=1.0)

COLOR_AUG_ONEOF = build_color_aug(WEIGHTS)

def build_pipeline(do_flip: bool) -> A.Compose:
    ops = []
    if do_flip:
        ops.append(A.HorizontalFlip(p=1.0))   # ê²°ì •ì  ì¢Œìš°ë°˜ì „
    ops.append(COLOR_AUG_ONEOF)               # ìƒ‰ìƒ ì¦ê°•ì€ í•­ìƒ í¬í•¨
    return A.Compose(
        ops,
        bbox_params=A.BboxParams(
            format="yolo",
            label_fields=["class_labels"],
            min_visibility=0.0
        )
    )

# ===== ë¼ë²¨ I/O =====
def read_yolo_label(lbl_path: Path):
    boxes, cls = [], []
    if not lbl_path.exists():
        return boxes, cls
    with open(lbl_path, "r") as f:
        for line in f:
            p = line.strip().split()
            if len(p) != 5:
                continue
            c = int(float(p[0]))
            x, y, w, h = map(float, p[1:5])
            boxes.append([x, y, w, h]); cls.append(c)
    return boxes, cls

def write_yolo_label(lbl_path: Path, boxes, cls):
    lbl_path.parent.mkdir(parents=True, exist_ok=True)
    with open(lbl_path, "w") as f:
        for (x, y, w, h), c in zip(boxes, cls):
            x = min(max(x, 0.0), 1.0); y = min(max(y, 0.0), 1.0)
            w = min(max(w, 0.0), 1.0); h = min(max(h, 0.0), 1.0)
            if w <= 0.0 or h <= 0.0:
                continue
            f.write(f"{int(c)} {x:.6f} {y:.6f} {w:.6f} {h:.6f}\n")
    # boxesê°€ ë¹„ì–´ë„ íŒŒì¼ì€ ìƒì„±ë¨(ë¹ˆ ë¼ë²¨)

# ===== ë°•ìŠ¤ ì •ê·œí™”(ì—ëŸ¬ ë°©ì§€) =====
def sanitize_yolo_boxes(boxes, cls, eps=1e-6):
    sane_boxes, sane_cls = [], []
    for (x, y, w, h), c in zip(boxes, cls):
        x = float(min(max(x, 0.0), 1.0))
        y = float(min(max(y, 0.0), 1.0))
        w = float(min(max(w, 0.0), 1.0))
        h = float(min(max(h, 0.0), 1.0))

        if w <= eps or h <= eps:
            continue

        half_w, half_h = w / 2.0, h / 2.0
        # ì¤‘ì‹¬ ìœ„ì¹˜ë¥¼ ì¡°ì •í•´ ë°•ìŠ¤ê°€ [0,1] ì•ˆì— ë“¤ë„ë¡
        if x - half_w < 0.0: x = half_w
        if x + half_w > 1.0: x = 1.0 - half_w
        if y - half_h < 0.0: y = half_h
        if y + half_h > 1.0: y = 1.0 - half_h

        if w <= eps or h <= eps:
            continue

        sane_boxes.append([x, y, w, h])
        sane_cls.append(c)
    return sane_boxes, sane_cls

# ===== ì‹¤í–‰ ì‹œì‘ =====
ensure_dirs()
print(f"ğŸ“ ê²°ê³¼ ë£¨íŠ¸ í´ë”: {OUT_DIR}")

summary = {
    "train": {"img": 0, "lbl": 0, "mode": ""},
    "val":   {"img": 0, "lbl": 0, "mode": ""},
    "test":  {"img": 0, "lbl": 0, "mode": "orig+aug_divot_priority"},
}

# 1) train/val: ì›ë³¸ ê·¸ëŒ€ë¡œ í¬í•¨ (ì‹¬ë³¼ë¦­ ë§í¬ ì‹œë„ â†’ ì‹¤íŒ¨í•˜ë©´ ë³µì‚¬)
for split in ("train", "val"):
    src_i, src_l = SRC_IMG[split], SRC_LBL[split]
    dst_i, dst_l = DST_IMG[split], DST_LBL[split]

    linked = False
    if USE_SYMLINKS and src_i.exists() and src_l.exists():
        # ëª©ì ì§€ í´ë” ì´ˆê¸°í™”
        if dst_i.exists() and any(dst_i.iterdir()):
            shutil.rmtree(dst_i)
            dst_i.mkdir(parents=True, exist_ok=True)
        if dst_l.exists() and any(dst_l.iterdir()):
            shutil.rmtree(dst_l)
            dst_l.mkdir(parents=True, exist_ok=True)

        linked_ok = True
        li = ll = 0
        for p in (src_i.iterdir() if src_i.exists() else []):
            if p.suffix.lower() in IMG_EXTS:
                if not safe_symlink(p, dst_i / p.name):
                    linked_ok = False
                    break
                li += 1
        for p in (src_l.iterdir() if src_l.exists() else []):
            if p.suffix.lower() == ".txt":
                if not safe_symlink(p, dst_l / p.name):
                    linked_ok = False
                    break
                ll += 1

        if linked_ok:
            linked = True
            summary[split]["img"] = li
            summary[split]["lbl"] = ll
            summary[split]["mode"] = "symlink"

    if not linked:
        n_img, n_lbl = copy_dir_tree_images_labels(src_i, src_l, dst_i, dst_l)
        summary[split]["img"] = n_img
        summary[split]["lbl"] = n_lbl
        summary[split]["mode"] = "copy"

    print(f"âœ… {split}: {summary[split]['mode']} (images={summary[split]['img']}, labels={summary[split]['lbl']})")

# 2) test: ì›ë³¸ ë³µì‚¬ + ë””ë´‡ ìœ„ì£¼ ìƒ‰ìƒ ì¦ê°•(ì¢Œìš°ë°˜ì „ í™•ë¥ ì ), ì´ ê°œìˆ˜ ì œí•œ
src_i, src_l = SRC_IMG["test"], SRC_LBL["test"]
dst_i, dst_l = DST_IMG["test"], DST_LBL["test"]

# --- test ì´ë¯¸ì§€ ë©”íƒ€ ì •ë³´ ìˆ˜ì§‘ (ì •ë ¬ + Divot í¬í•¨ ì—¬ë¶€) ---
img_infos = []
if src_i.exists():
    for img_path in sorted(src_i.iterdir()):
        if img_path.suffix.lower() not in IMG_EXTS:
            continue
        stem = img_path.stem
        lbl_path = src_l / f"{stem}.txt"
        has_divot = False
        if lbl_path.exists():
            try:
                with open(lbl_path, "r") as f:
                    for line in f:
                        parts = line.strip().split()
                        if len(parts) >= 1:
                            c = int(float(parts[0]))
                            if c == 0:   # Divot class
                                has_divot = True
                                break
            except Exception:
                pass
        img_infos.append({"img": img_path, "lbl": lbl_path, "has_divot": has_divot})

N_orig = len(img_infos)
print(f"\nğŸ§® test ì›ë³¸ ì´ë¯¸ì§€ ê°œìˆ˜: {N_orig}")

# --- case 1: ì›ë³¸ë§Œìœ¼ë¡œ ì´ë¯¸ MAX_TEST_IMAGES ì´ìƒì¸ ê²½ìš° ---
if N_orig >= MAX_TEST_IMAGES:
    print(f"âš ï¸ ì›ë³¸ test ì´ë¯¸ì§€ê°€ {N_orig}ì¥ìœ¼ë¡œ, MAX_TEST_IMAGES({MAX_TEST_IMAGES}) ì´ìƒì…ë‹ˆë‹¤.")
    print(f"   â†’ ì•ì—ì„œ {MAX_TEST_IMAGES}ì¥ë§Œ ë³µì‚¬í•˜ê³ , ì¦ê°•ì€ ìˆ˜í–‰í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
    used_infos = img_infos[:MAX_TEST_IMAGES]

    img_cnt = lbl_cnt = 0
    for info in used_infos:
        img_path = info["img"]
        lbl_path = info["lbl"]
        # ì´ë¯¸ì§€ ë³µì‚¬
        shutil.copy2(img_path, dst_i / img_path.name)
        img_cnt += 1
        # ë¼ë²¨ ë³µì‚¬ (ì¡´ì¬í•˜ëŠ” ê²½ìš°ë§Œ)
        if lbl_path.exists():
            shutil.copy2(lbl_path, dst_l / lbl_path.name)
            lbl_cnt += 1

    summary["test"]["img"] = img_cnt
    summary["test"]["lbl"] = lbl_cnt
    summary["test"]["mode"] = "orig_only(limit)"

else:
    # --- case 2: ì›ë³¸ì´ MAX_TEST_IMAGESë³´ë‹¤ ì ìœ¼ë©´, ëª¨ë‘ ë³µì‚¬ + ì¦ê°•ìœ¼ë¡œ ì±„ì›€ ---
    needed_aug = MAX_TEST_IMAGES - N_orig
    print(f"âœ… ì›ë³¸ {N_orig}ì¥ ë³µì‚¬ í›„, ì¦ê°•ìœ¼ë¡œ {needed_aug}ì¥ ì¶”ê°€ ì˜ˆì • (ìµœëŒ€ {MAX_TEST_IMAGES}ì¥ê¹Œì§€).")

    # 2-1) ì›ë³¸ ì´ë¯¸ì§€/ë¼ë²¨ ê·¸ëŒ€ë¡œ ë³µì‚¬
    img_cnt = lbl_cnt = 0
    for info in img_infos:
        img_path = info["img"]
        lbl_path = info["lbl"]
        shutil.copy2(img_path, dst_i / img_path.name)
        img_cnt += 1
        if lbl_path.exists():
            shutil.copy2(lbl_path, dst_l / lbl_path.name)
            lbl_cnt += 1
    summary["test"]["img"] = img_cnt
    summary["test"]["lbl"] = lbl_cnt

    # 2-2) ë””ë´‡ í¬í•¨ ì´ë¯¸ì§€/ê¸°íƒ€ ì´ë¯¸ì§€ ë¦¬ìŠ¤íŠ¸ ë¶„ë¦¬
    divot_infos = [info for info in img_infos if info["has_divot"]]
    other_infos = [info for info in img_infos if not info["has_divot"]]

    print(f"   Â· Divot í¬í•¨ ì´ë¯¸ì§€: {len(divot_infos)}ì¥")
    print(f"   Â· Divot ë¯¸í¬í•¨ ì´ë¯¸ì§€: {len(other_infos)}ì¥")

    # 2-3) ì¦ê°• ëŒ€ìƒ ì´ë¯¸ì§€ ì…€ë ‰ì…˜ (ë””ë´‡ ìš°ì„ )
    aug_candidates = []

    if needed_aug > 0:
        # ë¨¼ì € ë””ë´‡ í¬í•¨ ì´ë¯¸ì§€ì—ì„œ ì„ íƒ
        if len(divot_infos) >= needed_aug:
            aug_candidates = random.sample(divot_infos, needed_aug)
        else:
            aug_candidates = list(divot_infos)
            remain = needed_aug - len(divot_infos)
            if remain > 0 and len(other_infos) > 0:
                extra = random.sample(other_infos, min(remain, len(other_infos)))
                aug_candidates.extend(extra)

    print(f"   â†’ ì‹¤ì œ ì¦ê°• ëŒ€ìƒ ì´ë¯¸ì§€ ìˆ˜: {len(aug_candidates)}ì¥")

    # 2-4) ì„ íƒëœ ì´ë¯¸ì§€ë“¤ë§Œ 1íšŒì”© ì¦ê°• (ê°ê° _color ë˜ëŠ” _color_flip)
    aug_img_cnt = aug_lbl_cnt = 0
    for info in aug_candidates:
        img_path = info["img"]
        lbl_path = info["lbl"]

        # ë¼ë²¨ ì½ê¸° (ì—†ìœ¼ë©´ BG ì´ë¯¸ì§€)
        if lbl_path.exists():
            boxes, cls = read_yolo_label(lbl_path)
        else:
            boxes, cls = [], []

        # íŒŒì´í”„ë¼ì¸ êµ¬ì„± (flip ì—¬ë¶€ ëœë¤)
        do_flip = (random.random() < HFLIP_PROB)
        pipeline = build_pipeline(do_flip)

        # ì›ë³¸ ì´ë¯¸ì§€ ë¡œë”©
        img_bgr = cv2.imread(str(img_path))
        if img_bgr is None:
            print(f"âš ï¸ ì´ë¯¸ì§€ ë¡œë“œ ì‹¤íŒ¨(ì¦ê°• ìŠ¤í‚µ): {img_path}")
            continue
        img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

        # ë°•ìŠ¤ ì •ê·œí™”
        if len(boxes) > 0:
            boxes, cls = sanitize_yolo_boxes(boxes, cls)

        # ë³€í™˜ ì‹¤í–‰
        if len(boxes) == 0:
            transformed = pipeline(image=img_rgb, bboxes=[], class_labels=[])
            out_rgb = transformed["image"]
            out_bgr = cv2.cvtColor(out_rgb, cv2.COLOR_RGB2BGR)
            out_boxes, out_cls = [], []
        else:
            transformed = pipeline(image=img_rgb, bboxes=boxes, class_labels=cls)
            out_rgb = transformed["image"]
            out_bgr = cv2.cvtColor(out_rgb, cv2.COLOR_RGB2BGR)
            out_boxes = transformed["bboxes"]
            out_cls   = transformed["class_labels"]

        # íŒŒì¼ëª… ì ‘ë¯¸ì–´ (ì›ë³¸ê³¼ ë‹¤ë¥¸ ì´ë¦„ â†’ ì •ë ¬ ì‹œ ì›ë³¸ ë‹¤ìŒì— ì˜´)
        suffix = "_color_flip" if do_flip else "_color"
        new_img_name = img_path.stem + suffix + img_path.suffix
        new_lbl_name = img_path.stem + suffix + ".txt"

        # ì €ì¥
        cv2.imwrite(str(dst_i / new_img_name), out_bgr)
        aug_img_cnt += 1
        write_yolo_label(dst_l / new_lbl_name, out_boxes, out_cls)
        aug_lbl_cnt += 1

    summary["test"]["img"] += aug_img_cnt
    summary["test"]["lbl"] += aug_lbl_cnt

    print(f"   â†’ ì¦ê°•ìœ¼ë¡œ ì¶”ê°€ëœ ì´ë¯¸ì§€: {aug_img_cnt}ì¥, ë¼ë²¨: {aug_lbl_cnt}ê°œ")

print("\n================= SUMMARY =================")
for sp in ("train", "val", "test"):
    mode = summary[sp]["mode"]
    print(f"{sp:>5}: images={summary[sp]['img']:>5}  labels={summary[sp]['lbl']:>5}  ({mode})")

# 3) data.yaml ìƒì„±
yaml_text = f"""# Auto-generated by color-aug script (Albumentations + optional HFlip @ p={HFLIP_PROB})
path: {OUT_DIR.as_posix()}
train: images/train
val: images/val
test: images/test

nc: {len(YOLO_NAMES)}
names: {YOLO_NAMES}
"""
with open(OUT_DIR / "data.yaml", "w", encoding="utf-8") as f:
    f.write(yaml_text)

print("\nğŸ§¾ data.yaml ìƒì„± ì™„ë£Œ:")
print(OUT_DIR / "data.yaml")
print("\nâœ… ì‘ì—… ì™„ë£Œ!")
```
